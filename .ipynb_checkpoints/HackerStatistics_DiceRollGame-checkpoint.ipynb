# Hacker Statistics Project
Rules: Imagine you are walking stairs and playing a game with your friend. 
You throw a dice 100 times. 

If it's 1 or 2, you go one step down. 

If it's 3, 4, or 5, you go one step up. 

If you throw a 6, you have to throw the die again and walk up the resulting number of steps.

There is a 0.1% chance of falling down the stair when you make a move, basically having sto start from step 0 again. You bet your friend that you will reach up to 60 steps. 

What is the chance you will win this bet?

```python
import numpy as np
```

```python
# Need random number generator for dice (set the seed)
np.random.seed(123) #consistent/reproducible random numbers from the same seed
print(np.random.rand()) #pseudo-random numbers
```

```python
# Demonstrating/testing randInt() to simulate a dice
print(np.random.randint(1,7)) # 7 is not included
```

```python
# starting step
step = 50
```

```python
# roll dice
dice = np.random.randint(1,7)
```

```python
# rules
if dice <= 2:
    step = step - 1
elif dice <= 5 and dice >= 3:
    step = step + 1
else:
    step = step + np.random.randint(1,7)
```

```python
print("Dice Roll: " + str(dice))
print("# of steps: " + str(step))
```

### Random Walk (for loop) & putting it all together

```python
# Initialize random_walk list
random_walk = [0]

for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        # to avoid negative numbers from showing up in list
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

### Visualization

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(random_walk)
plt.show()
```

```python
#clear plot
plt.clf()
```

### Distribution
Getting an idea about how big your chances are of reaching 60 steps: need to repeatedly simulate the random walk and collect the results

```python
# Initialize all_walks
all_walks = []

# Simulate random walk 10 times (10 sets of 100 dice rolls)
for i in range(10) :
    
    # Code from before
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks; list of lists
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

```python
#converting all_walks list to NumPy array
np_aw = np.array(all_walks)
```

```python
#visualizing np_aw
plt.figure(figsize=(10, 6))
plt.plot(np_aw)
plt.show()
```

```python
#transposing/switching x & y values
np_aw_t = np.transpose(np_aw)
```

```python
# Plot np_aw_t and show
plt.figure(figsize=(10, 6))
plt.plot(np_aw_t)
plt.show()
```

### 250 simulations 
### Implementing clumsiness (0.1% of falling down, back to step 0)

```python
all_walks = []

for i in range(250) :
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        clumsy = np.random.rand()
        if clumsy <= 0.001:
            step = 0

        random_walk.append(step)
        
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

### Finally, what are the odds that you'll reach 60 steps high on the Empire State Building? (Visualization)

What are the end points of all the random walks simulated? Need to perform distribution (histogram)

```python
# Simulate random walk 500 times
all_walks = []

for i in range(500) :
    random_walk = [0]
    
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        
        clumsy = np.random.rand()
        if clumsy <= 0.001:
            step = 0
            
        random_walk.append(step)
        
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t, call it 'ends'
#This contains the endpoint of all 500 random walks you've simulated.
# Gets last number (step) in each random_walks list for each trial
ends = np_aw_t[-1, :]

# Plot histogram of ends, display plot
plt.figure(figsize=(10, 6))
plt.hist(ends)
plt.show()
```

### Finally, what are the odds that you'll reach 60 steps high on the Empire State Building? (Percentage Odds)

To calculate the chance, you can count the number of integers in ends that are greater than or equal to 60 and divide that number by 500, the total number of simulations.

```python
count = 0

for i in ends:
    if i >= 60:
        count += 1
        
percentageOdds = (count / 500) * 100

print("Odds that you will reach 60 steps high: " + str(percentageOdds) + "%")
```
